# Conditional Filtering
Filter by single condition
Filter by multiple conditions
Check against multiple values

```python
#Filtering a single condition.
df['total_bill'] #grab column to filter
bool_series = df['total_bill'] > 40 #where is it greater that 40, returns booleans
bool_series

```

```python
#now can pass this into dataframe to use as a filter
df[bool_series]
```

```python
#or just in one step like this really
df[df['total_bill'] > 40]
```

```python
df[df['sex'] ==  'Male']  #only males
```

```python
df[df['tip'] < 1.20] #find the low tippers
```

Multiple conditions

```python
#AND & -- BOTH CONDTIONS MUST BE TRUE
# OR | --- EITHER CONDITIONS ARE TRUE
```

```python
df['total_bill'] > 30
df['sex'] == "Male" #two conditions
```

```python
df[(df['total_bill']> 30) & (df['sex'] == "Male")] #for AND both to be true
#note the brackets!
```

Filtering by a list of options

```python
df[(df['day'] == 'Sun') | (df['day'] == 'Sat')]
#so saturday OR sunday using the pipe |
```

isin Method

```python
#ISIN method true or false if values in list of options

options = ['Sat', 'Sun', 'Mon']
df['day'].isin(options)
df[df['day'].isin(options)]
```

# Apply Methods
Apply on Single Column

```python
# .apply() method call to apply any python function to every row in a series!
```

```python
df.head()
```

```python
#e.g. grabbing the last four digits of CC number -> which is an integer
#so we want to cast it to a string and then slice it
# e.g. like this
str(12345675634) [-4:]
```

```python
#imma make a function with it
def last_four(num):
  return str(num)[-4:] #return it back as the sliced string, but it could come back as int too
```

```python
last_four(12345675634)
```

```python
#get that function applied with apply() !!
df['CC Number'].apply(last_four) #BAM!
df['last_four'] = df['CC Number'].apply(last_four) #make it a column if you want
df['last_four']
```

```python
df['total_bill']
```

```python
#We are going to return $ , $$, $$$ based on the total bill!
df['total_bill'].mean()
```

```python
def yelp(price):
  if price < 10:
    return '$'
  elif price >= 10 and price < 30:
    return '$$'
  else:
    return '$$$'
    
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
df['yelp'] #BOOM!
```

# Apply on multiple columns

```python
#Using the lambda expressions e.g.
#lambda num : num*2

df['total_bill'].apply(lambda num : num*2) #multiply bill value by 2
```

```python
df.head()
```

```python
#use apply method on total bill and tip, so we can see ratio of tip:total bill, then
# mark as generous tip or not

def quality(total_bill, tip):
  if tip/total_bill > 0.25:
    return "Generous"
  else:
    return "Other"
```

```python
quality(15,5)
```

```python
#so we can pass total bill and tip using lambda to the quality funcion!
df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
#axis =1 referring to columns
```

np.vectorize

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
#or this is simpler really and much faster
df.tail()
```

```python
#np.vectorize is to transform functions that are not numpy aware i.e. our custom functions
#which are just standard python functions
```
