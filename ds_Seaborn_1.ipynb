# Seaborn: Scatter

```python
#r/ship beween two continuous features
#e.g. x axis is salary, y is sales amount
```

```python
#since it is using matplotlib can alter the canvas prior
plt.figure(figsize=(10,4), dpi=200)
sns.scatterplot(x='salary',y='sales', data=df) #x and y from columns in the dataframe
#that we pass to it i.e. df
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.scatterplot(x='salary',y='sales', data=df, hue='level of education', palette='Dark2')
#using hue on level of education column, color the points by this column
#you can choose your own color pallette if wanted. don't mess them since they work
#for people who are color blind
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.scatterplot(x='salary',y='sales', data=df, hue='salary')

#so hue can be from continuous as here or categorical as above !
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.scatterplot(x='salary',y='sales', data=df, size='salary')
#could change the size of the points based off salary instead of using hue
#s= the general size, if you want to change all point sizes
#alpha= transparency so e.g. 0.2 if you want to see stacked points better
#you can do style=columname but it looks a bit bad but needed if don't have color 

#saving
#plt.savefig('my_plot.png')
```

# Distribution Plots

1. Rug Plot
2. Histogram
3. KDE

```python
sns.rugplot(x='salary', data=df, height=0.5) #rugplot has no y
```

```python
sns.displot(data=df, x='salary', bins=10) #alter the bins as needed
#DISPLOT
```

```python
sns.set(style='darkgrid') #set a style, high level command, so everyone gets the style

sns.displot(data=df, x='salary', bins=100, color='purple')
```

```python
sns.histplot(data=df, x='salary') #histplot
```

```python
sns.displot(data=df, x='salary', bins=100, color='purple', kde=True)
#with KDE over the top
```

```python
sns.kdeplot(data=df, x='salary') #just KDE plot along
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200) #make random ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age']) #make df
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages, x='age', height=0.5) #rugplot
```

```python
#distribution plot
sns.displot(data=sample_ages, x='age', bins=30)
```

```python
sns.displot(data=sample_ages, x='age', bins=30, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x='age')
#so there is some probability outside our range
#you can cut this off with clip=[0,100] etc.
```

```python
#we can edit the bandwidth with bw_adjust
sns.kdeplot(data=sample_ages, x='age', bw_adjust=0.1, shade=True)
#if it's too big it's just a normal distribution !
```

# Categorical Plots

```python
#countplot counts rows per category
#barplot any metric for the category
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.countplot(data=df, x='division')
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.countplot(data=df, x='level of education')
```

```python
#can add a hue to this too!
plt.figure(figsize=(10,4), dpi=150)
sns.countplot(data=df, x='level of education', hue='division')
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd',
            hue='division')

plt.legend(bbox_to_anchor=(1.05,1)) #get it outside the plot

#after grouping, calc the mean for y column per category
#ci confidence interval = standard deviation sd
```

```python
#some continuous and categorical data here
plt.figure(figsize=(10,4), dpi=150)
sns.boxplot(data=df, y='math score', x='test preparation course')
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.boxplot(data=df, y='math score', x='lunch')
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.boxplot(data=df, y='math score', x='parental level of education')
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.boxplot(data=df, y='reading score', x='lunch')
```

```python
#look at reading score and lunch with hue of test prep course
plt.figure(figsize=(10,4), dpi=150)
sns.boxplot(data=df, y='math score', x='lunch', hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,8), dpi=150) #a new orientation
sns.boxplot(data=df, x='math score', y='parental level of education', hue='test preparation course')
#no idea what non completed is from the graph tho
#see e.g. https://matplotlib.org/stable/tutorials/intermediate/legend_guide.html
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,8), dpi=150) #a new orientation
sns.violinplot(data=df, x='math score', y='parental level of education') 
#now with the violin plot which is mirrored KDE, can have hue if wanted still
```

```python
#SWARM PLOT KDE but with all the points shown
sns.swarmplot(data=df, x='math score')
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.swarmplot(data=df, x='math score', y='gender', size=2)
```

```python
#adding hue which isn't much fun to read
plt.figure(figsize=(10,4), dpi=150)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course')
plt.legend(bbox_to_anchor=(1.25,1))
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course', dodge=True)
plt.legend(bbox_to_anchor=(1.25,1))
#hue, but split them up
```

```python
#The BoxenPlot
plt.figure(figsize=(10,4), dpi=150)
sns.boxenplot(x='math score', y='test preparation course', hue='gender', data=df)
```

# Grid Plots

```python
#create a grid off of a categorical column, pass in name of column
#sns.catplot(x=,y=, row=, col=, kind=)
plt.figure(figsize=(10,4), dpi=150)
sns.catplot(data=df, x='gender', y='math score', kind='box')
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch', row='test preparation course')
```

```python
g= sns.PairGrid(df, hue='gender') #DIY #could add hue here

g= g.map_upper(sns.scatterplot)
g= g.map_lower(sns.kdeplot)
g= g.map_diag(sns.histplot)

g= g.add_legend()
```
