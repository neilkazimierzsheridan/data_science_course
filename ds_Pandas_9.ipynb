# Pivot Tables

```python
#a dataframe with repeated values can be pivoted for a reorganization and clarity
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['Sales_Funnel_CRM.csv'])) #remember can read in lots of other types too!
```

```python
df
```

```python
#how many licenses of each product type did google purchase? So we want company,
#number of licenses and product for each one
licenses = df[['Company', 'Product','Licenses']]
```

```python
licenses
```

```python
#time to pivot
pd.pivot(data=licenses, index='Company', columns='Product', values='Licenses')
```

```python
#an aggregation call of sum of licenses and sum of sales price
pd.pivot_table(df, index="Company", aggfunc="sum") #sum every column which might be a bit wacky
```

```python
results = df.groupby('Company').sum() #alternatively
```

```python
results
```

```python
#let's select columns we do want to use aggregate function on
pd.pivot_table(df, index="Company", aggfunc="sum", values=["Licenses", "Sale Price"])
```

```python
#two columns for the index (account manager is outer index, contact is inner)
pd.pivot_table(df, index=['Account Manager', 'Contact'], values=['Sale Price'], 
               columns=['Product'], aggfunc=[np.sum,np.mean])
```
